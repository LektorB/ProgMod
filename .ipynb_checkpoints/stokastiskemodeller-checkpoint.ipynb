```python
# eksempel 7.2
def kast(): 
    """ returner antall kast før det slåes seks på en terning"""
    antall_kast=0
    terning = 0
    
    while terning !=6:
        terning = np.random.randint(1,7)
        antall_kast += 1
    
    return antall_kast

# læper gjennom kastforsøket N ganger
N = 100000
L=[]
for i in range(N):
    L.append(kast())
    
print(f'Forventningsverdien er lik gjennomsnittet av  alle forsøkene i L: {np.mean(L)}')

#Regner ut glidende gjennomsitt:
nummer_i_rekken=np.arange(1,len(L)+1)
glidende_gjennomsnitt=np.cumsum(L)/nummer_i_rekken

#Plotter resultatet 
plt.plot(nummer_i_rekken,glidende_gjennomsnitt)
plt.xlabel('Antall forsøk')
plt.ylabel('Glidende gjennomsnitt')
plt.show()
```

```python
#7.5
# Lager en funksjon som velger tre biler og som returnerer true om 
# Nilsen blir hentet med samme bilmerke 3 ganger i løpet av uken

def B():
    D = ["Opel", "Volvo", "Ford", "Honda"]
    d = np.random.choice(D, 5, p = [0.30, 0.15, 0.35, 0.20])
    d = list(d) # koverterer til liste for å kunne bruke liste metoden count()  
    frekvens = [d.count(x) for x in D]
    return 3 in frekvens # dersom en av frekvensene er 3 retuneres True

N = 10000
suksess = [B() for x in range(N)]
gunstige = sum(suksess)

print(f'Sannsynligheten for å bli hentet med samme bilmerke tre ganger '
      f'samme uke er {gunstige/N}')
```

```python
# 7.6

def blod():
    """
    Trekker tilfeldig 2 personer, og sjekker om person 1 
    kan gi blod til person 2
    """ 
    type = ['AB','A','B','0']
    prob = [0.04,0.48,0.08,0.40]
    pers1 = np.random.choice(type,p=prob)
    pers2 = np.random.choice(type,p=prob)
    
    if pers1 == 'AB':
        return pers2 == 'AB'
    elif pers1 == 'B':
        return pers2 == 'AB' or pers2 == 'B'
    elif pers1 == 'A':
        return pers2 == 'AB' or pers2 == 'A'    
    else: 
        return pers1 == '0'
    
# kjører N - forsøk
N = 10000
suksess = [blod() for x in range(N)]

print(
      f'Sannsynligheten for at en til feldig person kan'
      f'gi blod til \n neste tilfeldige person er {sum(suksess)/N}'
    )
    
```

```python
# fyller baller i krukkene
rod = ['R']
bla = ['B']
krukke1 = 4*rod + 5*bla
krukke2 = 8*rod + 4*bla

def fire_rode():
    """ trekker tilsammen 7 kuler og sjekker om det er flere enn 4 rode"""
    trekk = (
            list(np.random.choice(krukke1,3,replace = False)) 
            + list(np.random.choice(krukke2,4,replace = False))
            )
    return trekk.count('R') > 4

N = 10000
suksess = [fire_rode() for x in range(N)]

print(f'Sannsynligheten for å trekke mer en 4 røde kuler er {sum(suksess)/N}')
```


Under er en mulig løsning hvor vi velger vinkel vilkårlig. Vi velger en sirkel med radius 1, da kan det vises at trekanten har sidekanter lik $\sqrt{3}$

```python
N=1000000
vinkel1 = np.random.uniform(0,2*np.pi,N)
vinkel2 = np.random.uniform(0,2*np.pi,N)

lengde = np.sqrt((np.cos(vinkel1)-np.cos(vinkel2))**2
                 + (np.sin(vinkel1)-np.sin(vinkel2))**2)

gunstige=sum(lengde>np.sqrt(3))

print(gunstige/N)
```

```python
# 1 og 2
N = 100000
sider = 20
def kast_alle():
    i=0
    a={s for s in range(1,sider+1)}
    
    while len(a) != 0: 
        i += 1
        kast = np.random.randint(1,sider + 1)
        
        if kast in a:
            a.remove(kast)
    return i

Liste = []

for i in range(N):
    Liste.append(kast_alle())

plt.hist(Liste)
print(f'median er {np.median(Liste)}, det betyr at i minst 50% av tilfellene brukte man {np.median(Liste)} eller færre kast på å få alle mulighetene')
```

```python
# 3, Denne var treig!

N = 10000
def kast_mynt():
    s = ['kron','mynt']
    antall = 0
    i = 0
    
    while antall < 10:
        i += 1
        kast = np.random.choice(s)
        
        if kast == 'kron':
            antall +=1
        else:
            antall = 0
    return i

myntkast=[]

for i in range(N):
    myntkast.append(kast_mynt())
    
plt.hist(myntkast)
print(np.median(myntkast))
```

```python
N = 100000
def sjanse(): 
    totalt = 0
    kast = 3
    terninger = 5
    
    while kast > 0 and terninger > 0:
        haand = np.random.randint(1,7,terninger)
        
        if kast == 1:
            totalt += sum(haand)
        
        else:
            seksere = sum(haand==6)
            terninger -= seksere
            totalt = totalt + seksere*6
        kast -= 1
    return totalt
            
Lsjanse = []

for i in range(N):
    Lsjanse.append(sjanse())

plt.hist(Lsjanse,bins=np.arange(4.5,31.5,1))
print(f'forventningsverdien er gjennosmnittet av summene: {np.mean(Lsjanse)}') 


# oppgave 6
print(f'Andel ganger over 15 poeng : { 100*sum(np.array(Lsjanse)>=15)/N:.2f}%')
print(f'Andel ganger over 20 poeng : { 100*sum(np.array(Lsjanse)>=20)/N:.2f}%')
print(f'Andel ganger over 25 poeng : { 100*sum(np.array(Lsjanse)>=25)/N:.2f}%')
```

```python
N = 100000
def sjanse(): 
    totalt = 0
    kast = 3
    terninger = 5
    
    while kast > 0 and terninger > 0:
        haand = np.random.randint(1,7,terninger)
        
        if kast == 1:
            totalt += sum(haand)
        
        elif kast == 3:
            seksere = sum(haand==6)
            terninger -= seksere
            totalt = totalt + seksere*6
        else:
            seksere = sum(haand == 6)
            femmere = sum(haand == 5)    
            terninger -= (seksere + femmere)
            totalt = totalt + seksere*6 + femmere*5
        kast -= 1
    return totalt
            
Lsjanse = []

for i in range(N):
    Lsjanse.append(sjanse())

plt.hist(Lsjanse,bins=np.arange(4.5,31.5,1))
print(f'forventningsverdien er gjennosmnittet av summene: {np.mean(Lsjanse)}') 

# oppgave 6
print(f'Andel ganger over 15 poeng : { 100*sum(np.array(Lsjanse)>=15)/N:.2f}%')
print(f'Andel ganger over 20 poeng : { 100*sum(np.array(Lsjanse)>=20)/N:.2f}%')
print(f'Andel ganger over 25 poeng : { 100*sum(np.array(Lsjanse)>=25)/N:.2f}%')
```
