```python
import numpy as np
import matplotlib.pyplot as plt



def eksakt(x):
    return 10071 * np.exp(-0.007*x) - 3571

def dif(N):
    return (-0.007*N-25)

N_0 = 6500
delta_t = 0.01

tid = np.arange(0,50+delta_t,delta_t)
N = np.zeros(len(tid))
N[0]=N_0

for i in range(1,len(N)):
    N[i] = N[i-1]+dif(N[i-1])*delta_t    


xverdier = np.linspace(0,50,300)
yverdier = eksakt(xverdier)
plt.plot(xverdier,yverdier,'--');
plt.plot(tid,N)
```

```python
import numpy as np
import matplotlib.pyplot as plt

#definerer området som skal tegnes og hvor langt det mellom pilene
Max=5
Min=-5
Step=0.4

#Her kan legge inn høyresiden av en diff-likingen av typen f'(x,y)
def dydx(x,y):
    try:
        return (np.exp(x)+2)/(2*y)
    except ZeroDivisionError:
        return np.NaN

xdot=np.arange(Min,Max,Step)
ydot=np.arange(Min,Max,Step)

#Lager "en matrise" 
X,Y=np.meshgrid(xdot,ydot)

dx=Step/2 # gjør dx litt mindre enn steg slik at ikke pilene overlapper
dy=dydx(X,Y)*dx

#normaliserer pilene slik at alle blir like lange:
N=np.sqrt(dx**2+dy**2)

dx=dx/N
dy=dy/N

#tegner piler fra X,Y til X+dx,Y+dy, litt samme syntax som i geogebra.
plt.quiver(X,Y,dx,dy,angles="xy")

#integralkurve fra x0 og y0:
XX=-4
YY=+0.1
Steg=0.01
xliste=[XX]
yliste=[YY]

while XX < Max and YY<Max and YY>Min:
    XX+=Steg
    YY+=dydx(XX,YY)*Steg
    xliste.append(XX)
    yliste.append(YY)
plt.plot(xliste,yliste,"-r")

```
