```python
x = 1

def f(y):        # inputen gies til den lokale variablen y
    x = 2*y      # det beregnes at den lokal variable x = 2
    print(2*x)   # det skrives ut 4 (altaså 2*2)
    
                # Funksjonen retunerer ingen verdier, de lokale variablene slettes
                # De globale variablene (x=1) forblir uendret
    
f(x)            # funksjonen kjører med den globale variablen x = 1 (4 skrives ut) 
print(x)        # Den globale variablen x skrives ut, altså 1
```

```python
def f(x):
    y = 0
    while x > 0:
        y = y + x % 10 
        x = int(x/10)
    return y
print(f(32145))
```

```python
def f(x):
    y = 0
    while x > 0:
        y = y + x % 10 
        x = int(x/10)
        
    if y >= 10:
        y = f(y)
        
    return y
print(f(32145))
```

Vi ser at det første programmet retunerer tversummen. Inne i funksjonen starter vi med y = 0 og så legger vi til ett og ett siffer i løkken. Vi starter bakerst. x%10 gir resten når vi deler på ti, det tilsvarer at vi deler på ti og ser hva som står bak kommet, som igjen er det samme som siste siffer. i neste linje deler vi tallet på ti, dropper alle desimaler og fortsetter i løkken til vi har fått med oss alle sifferne.

De to linjene som var kommentert bort i b) sjekker om tversummen var en-sifret. Dersom tversummen har flere siffer finner man tversummen av dette svaret også helt til man sitter igjen med et ettsifferet tall.


```python
def heron(a,b,c):
    from math import sqrt
    s = (1/2)*(a + b + c)
    radikand = s*(s-a)*(s-b)*(s-c)
    
    if radikand <= 0:
        print('Disse lengdene gir ikke en trekant')
    else:
        print(f'Trekant ABC har areal: {sqrt(radikand)}')    
                              
```

```python
heron(1,2,3)
```

```python
heron(3,3,5)
```

```python
def collatz(n):
    teller = 0
    while n>1:
        if n%2==0:
            n = n/2
        else:
            n = 3*n + 1
        teller += 1
    return int(n) , teller #retunerer n for sjekke at den faktisk er 1 og antall steg   
```

```python
# tester med kjente tall:
collatz(5)
```

```python
import numpy as np
antall_steg =[]

for i in range(2,1000001): # tester alle tall mellom 2 og 1 mill.
    en, steg =  collatz(i) 
    antall_steg.append(steg) # legger til alle stegene i liste


print(f'Den lengste stien er {np.max(antall_steg)} steg lang og '
      f'tilhører n = {np.argmax(antall_steg)+2}')
```

```python
import numpy as np
rng = np.random.default_rng()


def kast(n,N):
    terning1 = rng.integers(1,n+1,N)
    terning2 = rng.integers(1,n+1,N)
    return np.sum(terning1*terning2>60)/N
```

```python
# a) kjøre simuleringen med 1000000 kast et par ganger, ser at svaret svinger rundt 0.13
kast(10,1000000)
```

```python
# b)
n = 10 # vet at n = 10 gir mindre sannsynlighet enn 0.5, starter der.
N = 1000000
while kast(n,N) < 0.5:
    n += 1
    
print(f'Den minste terningen med P(x>60) > 0.5 har {n} sider og P(x>60) = {kast(n,N)}') 
```

```python
from numpy import exp
def f(x):
    return 5*x*exp(-2*x)
    
def derivert(x,h):
    return((f(x+h)-f(x-h))/(2*h))

def R(x,h):
    return (4*derivert(x,h/2) - derivert(x,h))/3

def analytisk(x):
    return (5*exp(-2*x)-10*x*exp(-2*x))
```

```python
# a)
print(f"f'(0.35)={derivert(0.35,0.25):.4f} når h = 0.25")
print(f"f'(0.35)={derivert(0.35,0.125):.4f} når h = 0.125")
```

```python
# c)
print(f'Sentral kvotient med h = 0.25 gir feilen:'
      f'{abs(analytisk(0.35)-derivert(0.35,0.25))/analytisk(0.35):.4f}')

print(f'Sentral kvotient med h = 0.125 gir feilen:'
      f'{abs(analytisk(0.35)-derivert(0.35,0.125))/analytisk(0.35):.4f}')

print(f'Richardson interpolasjon med h = 0.25 og h = 0.125 gir feilen:', 
      f'{abs(analytisk(0.35)-R(0.35,0.25))/analytisk(0.35):.4f}')
```
